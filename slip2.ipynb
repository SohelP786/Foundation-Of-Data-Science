```python
#Write a Python program for Handling Missing Value. Replace missing value of salary, 
#age column with mean of that column.(Use Data.csv file)
import pandas as pd
file=pd.read_csv('E:/Data.csv')
file

#replace missing value with their mean
#salery
file['salery']=file[salery].fillno(file['salery'].mean())

#age
file[age]=file[age].fileno(file[age].mean)
```

```python
#) Write a Python program to generate a line plot of name Vs salary
import matplotlib.pyplot as mt
file.plot(x='name',y='salery')
mt.show
```
