```python
#Write a program in python to perform following task : [15]
#Standardizing Data (transform them into a standard Gaussian distribution with a mean 
#of 0 and a standard deviation of 1) (Use winequality-red.csv)

#standerizing the data
import numpy as np
import pandas as pd
from sklearn import preprocessing
import scipy.stats as s
print("standerising the data")
print("---------------------------------------------------------")
x=pd.read_csv("E:/winequality-red.csv")
#x
print("original data\n",x)
print("\nInitial mean:",s.tmean(x).round(2))
print("\nInitial standerd daviation is:\n",round(x.std(),2))

x_scaled=preprocessing.scale(x)
x_scaled.mean(axis=0)
x_scaled.std(axis=0)
print("---------------------------------------------------------")
print("\nStanderized data\n",x_scaled.round(2))
print("\nscaled mean\n",s.tmean(x_scaled).round(2))
print("\nscaled standerd deviation\n",round(x_scaled.std(),2))
```
