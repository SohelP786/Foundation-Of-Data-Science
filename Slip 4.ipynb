```python
#Generate a random array of 50 integers and display them using a line chart, scatter 
#plot, histogram and box plot. Apply appropriate color, labels and styling options.
import numpy as np
import matplotlib.pyplot as plt
from numpy import random
x=np.random.rand(50)
y=np.random.rand(50)
plt.plot(x,y,color="r",linestyle=':')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
#histogram
plt.xlabel("X-axis")
plt.ylabel("y-axis")
plt.hist(x,color="r")
plt.show()
```

```python
#scatterplot
plt.scatter(x,y)
plt.show()
```

```python
#boxplot
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.boxplot(x)
plt.show()
```
